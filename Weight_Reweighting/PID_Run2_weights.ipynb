## Photon ID Run 2 weights

* `mcid` = dataset-id

* `mcWeight` = MC generator weight. 1 for LO generator, can be negative for NLO or higher (counterterms, virtual corrections, ...)

* `xs` = cross-section

* `kf` = k factor

* `ge` = generator efficiency (important in case a filter is applied at generator level)

* `intLumi` = integrated luminosity for corresponding periods (a, d,e)

* `nEvents` = number of generated events

* `evtWeight` = weight accouting for experimental effects: `evtWeight= puWeight * vtxWeight * yWeight`

* `puWeight` = pileup weight, accounts for differnce in $\mu$ distribution in data and MC

* `vtxWeight` = weight accounting for beamspot difference in data and MC

* `yWeight` = weight including data/MC scale factor to account for photon efficiency difference after FF

* `sumWeights` = sum of all weights in sample. Equal to number of generated events for LO samples

* `mcTotWeight` = weight accounting for all effects, when applied returns distribution normalised to integrated distribution in period.

    * `mcTotWeight = evtWeight * xs * ge * mcWeight * intLumi / sumWeights = evtWeight * lumiXsecWeight`
